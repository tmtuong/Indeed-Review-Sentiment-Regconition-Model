```python
import pygsheets
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
stars = []
body = []
good = []
bad = []
page = 0
driver = webdriver.Chrome()
while True:
    driver.get(f'https://www.indeed.com/cmp/Tesla/reviews?start={page*20}')
    current_link = driver.current_url
    if page >0 and str(current_link)[-7:] == 'reviews':
        break
    else:
        text = driver.page_source
        soup = BeautifulSoup(text)
        data = soup.find_all('div',class_='css-t3vkys eu4oa1w0')
        for raw in data:
            star = raw.find('div', class_ ='css-1h3aion eu4oa1w0').text[:3]
            stars.append(star)

            main = raw.find_all('span',class_ ='css-82l4gy eu4oa1w0')
            content = '\n'.join([i.text for i in main])
            body.append(content)
            
            help = raw.find_all('span',class_='css-1ot3ci0 eu4oa1w0')
            is_good = False
            is_bad = False


            for i in help:
                lst = i.text.split(' ')
                if 'unhelpful' in lst:
                    bad.append(int(lst[2]))
                    is_bad = True
                if 'helpful' in lst:
                    good.append(int(lst[2]))
                    is_good = True
            if is_good == False:
                good.append(0)
            if is_bad == False:
                bad.append(0)

    print('Page:',page+1)
    print('\n')
    page = page +1
```

```python
df = pd.DataFrame({'body':body,'star':stars,'good':good,'bad':bad})
df.drop_duplicates(inplace=True)
```
